# Linear Regression Example 

```python
# import everything we need
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('OmniPower.csv')
print(df.head())
```

```python
X = df[['Price', 'Promotion']]  
y = df['Sales']  
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

print(X_train.shape)
print(y_train.shape)

X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
linReg.fit(X_train, y_train)
```

```python
print(linReg.intercept_)
print(linReg.coef_)
```

```python
from sklearn.metrics import r2_score
y_train_pred = linReg.predict(X_train)
y_test_pred = linReg.predict(X_test)
print('r2_score: ',r2_score(y_test, y_test_pred)) 
# Compare the value with linReg.score().  Is the score the same as model.score?
print('score: ',linReg.score(X_test,y_test))
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_train.iloc[:,0], y_train)
plt.plot(X_train.iloc[:,0], y_train_pred, color='red')
plt.show()
```
